```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
lmbd = np.linspace(300,2000,1700,endpoint=True)
d = float(input('Schichtdicke in nm:'))
n0 = 1
n1 = float(input('Brechungsindex n1 (schicht): '))
n2 = float(input('Brechungsindex n2 (substrat): '))

r01 = (n0-n1)**2/(n0+n1)**2 

r12 = (n1-n2)**2/(n1+n2)**2

if n1>n2:
    delta = (2*np.pi)/(lmbd*10**-9) * n1 * d*10**-9 + np.pi
else:
    delta = (2*np.pi)/(lmbd*10**-9) * n1 * d*10**-9

R = (r01**2 + r12**2 + 2*r01*r12*np.cos(2*delta))/(1 + (r01*r12)**2 + 2*r01*r12*np.cos(2*delta))

    
plt.figure(figsize=(20,10))
plt.plot(lmbd,R)
plt.ylabel('R')
plt.xlabel('lambda')
plt.xticks(np.arange(300, 2100, step=100))
plt.grid()
plt.show()



    
```
